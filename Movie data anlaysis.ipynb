```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
```

```python
df['Release_Date'] = df['Release_Date'].dt.year
```

```python
df
```

```python
df.drop(['Overview', 'Original_Language', 'Poster_Url'], axis=1, inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
bins = [df['Vote_Average'].min(), df['Vote_Average'].quantile(0.25), df['Vote_Average'].quantile(0.50), df['Vote_Average'].quantile(0.75), df['Vote_Average'].max()]
labels = ['not_popular', 'below_average', 'average', 'popular']

df['Vote_Average'] = pd.cut(x=df['Vote_Average'], bins=bins, labels=labels, include_lowest=True, duplicates='drop')
```

```python
df['Vote_Average'].unique()
```

```python
df['Vote_Average'].value_counts()
```

```python
df.head()
```

```python
df.dropna(inplace=True)

df.isna().sum()
```

```python
df.info()
```

- we'd split genres into a list and than explode our dataframe to have only one genre per row for each movie

```python
df['Genre'] = df['Genre'].str.split(', ')

df = df.explode('Genre').reset_index(drop=True)
```

```python
df
```

#### Casting column into category

```python
df['Genre'] = df['Genre'].astype('category')

df['Genre'].dtypes
```
